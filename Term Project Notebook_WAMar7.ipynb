```python
import pandas as pd
import xlrd
import os
```

First thing, we download the weather information and we combine it!

```python
weather_archive = pd.read_excel("http://geog.utm.utoronto.ca/stats/utmms/UTMMS%20Full%20Data%20Jan%201%202000%20to%20Sept%2026%202018.xlsx",engine="xlrd")
weather_current = pd.read_csv("http://geog.utm.utoronto.ca/stats/utmms/UTMMS_Daily_Nov22_dbDump.csv")
weather = pd.concat([weather_archive,weather_current])
weather.to_csv("uft_weather_info.csv",sep=',')
#here I renamed the timestamp column to occurrencedate so that we could merge the two data sets into one using time.
weather = weather.rename(columns={'timestamp_est': 'occurrencedate'})
weather = weather.sort_values(by = 'occurrencedate')
weather.head(5)
```

```python
bne = pd.read_csv('bne.csv')
#changing index to event unique id
bne.set_index('event_unique_id', inplace=True)

#dropping X and Y as these two columns are in there twice; Long and Lat
bne.drop(['X', 'Y', 'Index_'], axis=1, inplace = True)

#sort by date
bne.sort_values(by= 'occurrencedate')

bne.head(5)
```

```python
#trying to merge data 
data = pd.merge(bne, weather)
data
```
